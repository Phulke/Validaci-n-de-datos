```python
import logging

# Configuraci√≥n de logging
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

def validar_datos(nombre, edad, correo, pais=None):

    logging.debug(f"Loading‚Ä¶ ‚ñà‚ñí‚ñí‚ñí‚ñí‚ñí‚ñí‚ñí‚ñí‚ñí Iniciando validaci√≥n: Nombre='{nombre}', Edad='{edad}', Correo='{correo}', Pa√≠s='{pais}'")

    es_valido = False

    try:
        #Verificaci√≥n TypeError

        if not isinstance(nombre, str) or not isinstance(correo, str):
            # 1. TypeError: Si el nombre o correo no son cadenas.
            raise TypeError("Error de tipo: Nombre y Correo deben ser cadenas de texto.")

        # Convertir la edad a entero. Si falla, genera ValueError/TypeError.
        try:
            edad_entera = int(edad)
        except (ValueError, TypeError):
            raise ValueError("Error de valor: La 'edad' debe ser un n√∫mero entero v√°lido.")


        #Verificaci√≥n de Contenido ValueError)

        if not nombre.strip():
            raise ValueError("Error de valor: El nombre no puede estar vac√≠o.")

        if edad_entera <= 0:
            #ValueError Si la edad no cumple que debe ser > 0
            raise ValueError("Error de valor: La edad debe ser mayor que cero.")

        if '@' not in correo:
            raise ValueError("Error de valor: El correo electr√≥nico debe contener el s√≠mbolo '@'.")


        # Verificaci√≥n de Campo Faltante KeyError
        # Simulamos que los datos se procesan desde un diccionario y 'pais' es un campo requerido.
        # En esta prueba, 'pais' es None (simulando que la clave est√° ausente o su valor es None,
        # pero se trata en este caso como un error de clave si no tiene valor).

        datos_registro = {"nombre": nombre, "edad": edad_entera, "correo": correo, "pais": pais}

        #KeyError: Lanzado si la clave pais est√° ausente o inv√°lida.
        # Se fuerza el error si el valor asociado a la clave es None podemos interpretarlo como  "falta el dato"
        if datos_registro["pais"] is None:
             raise KeyError("Error de clave: El campo 'Pa√≠s' es obligatorio y no puede estar vac√≠o.")


        # Si todas las validaciones son exitosas
        print(f"Ÿ©(^‚Äø^)€∂ √âxito: Datos de {nombre} ({datos_registro['pais']}) registrados correctamente.")
        es_valido = True

    #Captura de las 3 Excepciones Diferentes para realizar el ejercicio
    except TypeError as e:
        print(f"ùù¨ Error de Tipo (TypeError): {e}")
        logging.error(f"(‚ï•Ôπè‚ï•) Fallo en la validaci√≥n por tipo: {e}")

    except ValueError as e:
        print(f"ùù¨ Error de Valor (ValueError): {e}")
        logging.error(f"(‚ï•Ôπè‚ï•) Fallo en la validaci√≥n por valor/contenido: {e}")

    except KeyError as e:
        #Captura: Campo obligatorio (simulado como 'pais') no encontrado.
        print(f"ùù¨ Error de Clave (KeyError): {e}")
        logging.error(f"(‚ï•Ôπè‚ï•) Fallo por campo obligatorio faltante: {e}")

    except Exception as e:
        print(f"ùù¨ Error Desconocido: {e}")
        logging.error(f"(‚ï•Ôπè‚ï•) Error no manejado: {e}")

    finally:
        # Mensaje de cierre.
        print(f"„Éò( ^o^)„ÉéÔºº(^_^ )Cierre del proceso de validaci√≥n.\n{'-'*30}")
        return es_valido

def probar_validaciones():
    print(f"INICIANDO PRUEBAS DE VALIDACI√ìN DE DATOS. \n{'-'*30}")

    # 1. Prueba V√°lida (√âxito)
    validar_datos(" Ana L√≥pez", 28, "ana.lopez@gmail.com", pais="M√©xico")

    # 2. Prueba Inv√°lida: Nombre con Tipo Incorrecto (TypeError)
    print("Prueba Inv√°lida: Nombre con Tipo Incorrecto (TypeError)")
    validar_datos(12345, 22, "Maria12345@protonmail.com", pais="Chile")

    # 3. Prueba Inv√°lida: Nombre vacio (TypeError)
    print("Prueba Inv√°lida: Nombre vacio (TypeError)")
    validar_datos("", 22, "Milo@protonmail.com", pais="Argentina")

    # 4. Prueba Inv√°lida: Edad menor que 0 (ValueError)
    print("Prueba Inv√°lida: Edad menor que 0 (ValueError)")
    validar_datos("Elena D√≠az", 0, "elena@hotmail.com", pais="Per√∫")

    # 5. Prueba Inv√°lida: Correo sin @ (ValueError)
    print("Prueba Inv√°lida: Correo sin @ (ValueError)")
    validar_datos("Pedro Rojas", 30, "pedroejemplo.com", pais="Ecuador")

    # 6. Prueba Inv√°lida: Campo 'Pa√≠s' None (KeyError)
    print("Prueba Inv√°lida: Campo 'Pa√≠s' Faltante (KeyError)")
    # No se pasa el argumento 'pais', por lo que toma el valor predeterminado None, forzando el KeyError explicado anteriormente.
    validar_datos("Mia Smith", 35, "mia@tutanota.com")

#Ejecuta la funci√≥n de prueba
probar_validaciones()
```
